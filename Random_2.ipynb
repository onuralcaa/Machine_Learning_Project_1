```python
# Yılların tam aralığını belirleyin
tam_yillar = set(range(2000, 2020))  # 2000-2019 yılları

# Her ülkenin yıllarını kontrol edin ve eksiksiz olanları seçin
eksiksiz_ulkeler = df.groupby('Ulke')['Yil'].apply(set).apply(lambda x: x == tam_yillar)

# Eksiksiz veriye sahip ülkeleri filtreleyin
df = df[df['Ulke'].isin(eksiksiz_ulkeler[eksiksiz_ulkeler].index)]

# Temizlenmiş veri setini yeni bir dosyaya kaydedin
# temiz_df.to_csv("temizlenmis_veri_seti.csv", index=False)
```

```python
# Kalkınma endeksi hesaplama
df["Kalkinma_Endeksi"] = ((
    0.15 * df['Elektrik_Erisim_%'] +                      # Elektrik erişimi ağırlığı
    0.15 * df['Temiz_Yakit_Erisim'] +                     # Temiz yakıt erişimi ağırlığı
    0.15 * df['Yenilenebilir_Kapasite_KisiBasi'] +         # Yenilenebilir kapasite ağırlığı
    0.2 * df['Yenilenebilir_Enerji_Orani'] +              # Yenilenebilir enerji oranı ağırlığı
    0.1 * df['Gdp_KisiBasi'] +                            # Kişi başına GSYİH ağırlığı
    0.1 * df['Gdp_Buyume'] -                             # GSYİH büyüme oranı ağırlığı
    0.1 * df['Fosil_Yakit_Elektrik_TWh'] -                # Fosil yakıt elektriği ağırlığı
    0.05 * df['CO2_Emisyonu_kt']                          # CO2 emisyonu ağırlığı
) * -1 )

# Hesaplanan kalkınma endeksini kontrol etme
print(df[['Ulke', 'Yil', 'Kalkinma_Endeksi']].tail())

# Sonuçları CSV dosyasına kaydetme (opsiyonel)
# df.to_csv("kalkinma_endeksi_eklenmis.csv", index=False)
```

```python
# Her ülkenin ortalama kalkınma endeksini hesaplayın
ortalama_endeks = df.groupby('Ulke')['Kalkinma_Endeksi'].mean()

# En yüksek kalkınma endeksine sahip 8 ülkeyi seçin
en_iyi_8_ulke = ortalama_endeks.nlargest(8)

# Seçilen ülkelerin detaylarını filtreleyin
en_iyi_df = df[df['Ulke'].isin(en_iyi_8_ulke.index)]

# Sonuçları ekrana yazdırın
print(en_iyi_df[['Ulke', 'Yil', 'Kalkinma_Endeksi']].head(10))

# Opsiyonel: Sonuçları yeni bir CSV dosyasına kaydedin
# en_iyi_df.to_csv("en_iyi_8_ulke.csv", index=False)

print("En iyi kalkınma endeksine sahip 8 ülke başarıyla kaydedildi!")
```

## RANDOM FOREST VE XGBOOST MODELİ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

# Veriyi hazırlama
data = en_iyi_df_sade.copy()
train = data[data['Yil'] <= 2010]  # 2000-2010 eğitim seti
test = data[data['Yil'] > 2010]   # 2011-2019 test seti

# Performans sonuçlarını saklamak için bir sözlük
results = {}

# Her ülke için modeli çalıştır ve sonuçları ayrı grafiklerde çiz
for ulke in data['Ulke'].unique():
    # Ülke bazında filtreleme
    train_ulke = train[train['Ulke'] == ulke].copy()
    test_ulke = test[test['Ulke'] == ulke].copy()

    # -----------------------------------------
    # ÖZELLİK MÜHENDİSLİĞİ
    # -----------------------------------------

    # Gecikmeli (Lag) özellikler
    for lag in [1, 2, 3, 4, 5]:  # Daha fazla gecikmeli özellik
        train_ulke[f'Lag{lag}'] = train_ulke['Kalkinma_Endeksi'].shift(lag)
        test_ulke[f'Lag{lag}'] = train_ulke['Kalkinma_Endeksi'].iloc[-lag]

    # Fark (Diff) özellikleri
    train_ulke['Diff1'] = train_ulke['Kalkinma_Endeksi'].diff(1)
    test_ulke['Diff1'] = test_ulke['Kalkinma_Endeksi'].diff(1)

    # Hareketli ortalamalar (Moving Averages)
    train_ulke['MA3'] = train_ulke['Kalkinma_Endeksi'].rolling(window=3).mean()
    test_ulke['MA3'] = test_ulke['Kalkinma_Endeksi'].rolling(window=3).mean()

    train_ulke['MA5'] = train_ulke['Kalkinma_Endeksi'].rolling(window=5).mean()
    test_ulke['MA5'] = test_ulke['Kalkinma_Endeksi'].rolling(window=5).mean()

    # Mevsimsellik (Sin/Cos)
    train_ulke['sin_yil'] = np.sin(2 * np.pi * train_ulke['Yil'] / 12)
    train_ulke['cos_yil'] = np.cos(2 * np.pi * train_ulke['Yil'] / 12)

    test_ulke['sin_yil'] = np.sin(2 * np.pi * test_ulke['Yil'] / 12)
    test_ulke['cos_yil'] = np.cos(2 * np.pi * test_ulke['Yil'] / 12)

    # Eksik değerleri doldurma
    train_ulke = train_ulke.fillna(method='bfill').fillna(method='ffill')
    test_ulke = test_ulke.fillna(method='bfill').fillna(method='ffill')

    # -----------------------------------------
    # Özellik ve Hedef Değişkenler
    # -----------------------------------------
    features = ['Yil', 'Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 
                'Diff1', 'MA3', 'MA5', 'sin_yil', 'cos_yil']

    X_train = train_ulke[features]
    y_train = train_ulke['Kalkinma_Endeksi']
    X_test = test_ulke[features]
    y_test = test_ulke['Kalkinma_Endeksi']

    # -----------------------------------------
    # 1. Random Forest Modeli
    # -----------------------------------------
    rf_model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
    rf_model.fit(X_train, y_train)
    rf_preds = rf_model.predict(X_test)

    # Performans değerlendirme
    rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))
    rf_mae = mean_absolute_error(y_test, rf_preds)
    rf_r2 = r2_score(y_test, rf_preds)

    # -----------------------------------------
    # 2. XGBoost Modeli
    # -----------------------------------------
    xgb_model = XGBRegressor(
        n_estimators=500,         # Ağaç sayısı
        learning_rate=0.05,       # Öğrenme oranı
        max_depth=6,              # Derinlik artırıldı
        subsample=0.8,            # Rastgele örnekleme
        colsample_bytree=0.8,     # Özellik alt kümesi seçimi
        random_state=42
    )
    xgb_model.fit(X_train, y_train)
    xgb_preds = xgb_model.predict(X_test)

    # Performans değerlendirme
    xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_preds))
    xgb_mae = mean_absolute_error(y_test, xgb_preds)
    xgb_r2 = r2_score(y_test, xgb_preds)

    # -----------------------------------------
    # Sonuçları kaydet
    # -----------------------------------------
    results[ulke] = {
        'RF_RMSE': rf_rmse, 'RF_MAE': rf_mae, 'RF_R2': rf_r2,
        'XGB_RMSE': xgb_rmse, 'XGB_MAE': xgb_mae, 'XGB_R2': xgb_r2
    }

    # -----------------------------------------
    # Grafik çizimi
    # -----------------------------------------
    plt.figure(figsize=(10, 6))
    plt.plot(test_ulke['Yil'], y_test, label='Gerçek Değerler', color='blue', linestyle='--')
    plt.plot(test_ulke['Yil'], rf_preds, label='RF Tahmin', color='green', linestyle=':')
    plt.plot(test_ulke['Yil'], xgb_preds, label='XGB Tahmin', color='orange')
    plt.title(f'{ulke} Kalkınma Endeksi Tahmin Sonuçları')
    plt.xlabel('Yıl')
    plt.ylabel('Kalkınma Endeksi')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# -----------------------------------------
# Sonuçları tablo olarak göster
# -----------------------------------------
sonuclar_df = pd.DataFrame(results).T
print(sonuclar_df)
```

## SARIMA VE HOLT-WİNTERS MODELİ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Veriyi hazırlama
data = en_iyi_df_sade.copy()
train = data[data['Yil'] <= 2010]  # 2000-2010 eğitim seti
test = data[data['Yil'] > 2010]   # 2011-2019 test seti

# Performans sonuçlarını saklamak için bir sözlük
results = {}

# Her ülke için modeli çalıştır ve sonuçları ayrı grafiklerde çiz
for ulke in data['Ulke'].unique():
    # Ülke bazında filtreleme
    train_ulke = train[train['Ulke'] == ulke].copy()
    test_ulke = test[test['Ulke'] == ulke].copy()

    # Zaman serisi formatına geçiş
    train_ulke['Yil'] = pd.to_datetime(train_ulke['Yil'], format='%Y')
    test_ulke['Yil'] = pd.to_datetime(test_ulke['Yil'], format='%Y')
    train_ulke.set_index('Yil', inplace=True)
    test_ulke.set_index('Yil', inplace=True)

    # Hedef değişken
    y_train = train_ulke['Kalkinma_Endeksi']
    y_test = test_ulke['Kalkinma_Endeksi']

    # -----------------------------------------
    # 1. SARIMA Modeli
    # -----------------------------------------
    try:
        sarima_model = SARIMAX(y_train, 
                               order=(1, 1, 1), 
                               seasonal_order=(1, 1, 1, 3),  # Mevsimsellik 3 yıl olarak ayarlandı
                               enforce_stationarity=False, 
                               enforce_invertibility=False)
        sarima_fit = sarima_model.fit(disp=False)
        sarima_preds = sarima_fit.forecast(steps=len(y_test))
    except:
        sarima_preds = np.zeros(len(y_test))  # Hata olursa sıfır tahmini

    # SARIMA Performans değerlendirme
    sarima_rmse = np.sqrt(mean_squared_error(y_test, sarima_preds))
    sarima_mae = mean_absolute_error(y_test, sarima_preds)
    sarima_r2 = r2_score(y_test, sarima_preds)

    # -----------------------------------------
    # 2. Holt-Winters (ETS) Modeli
    # -----------------------------------------
    try:
        hw_model = ExponentialSmoothing(y_train, 
                                        trend='add',        # Trend: Artımsal
                                        seasonal='add',     # Mevsimsellik: Artımsal
                                        seasonal_periods=3) # Mevsimsel dönem uzunluğu: 3 yıl
        hw_fit = hw_model.fit(optimized=True)
        hw_preds = hw_fit.forecast(steps=len(y_test))
    except:
        hw_preds = np.zeros(len(y_test))  # Hata olursa sıfır tahmini

    # Holt-Winters Performans değerlendirme
    hw_rmse = np.sqrt(mean_squared_error(y_test, hw_preds))
    hw_mae = mean_absolute_error(y_test, hw_preds)
    hw_r2 = r2_score(y_test, hw_preds)

    # -----------------------------------------
    # Sonuçları kaydet
    # -----------------------------------------
    results[ulke] = {
        'SARIMA_RMSE': sarima_rmse, 'SARIMA_MAE': sarima_mae, 'SARIMA_R2': sarima_r2,
        'HW_RMSE': hw_rmse, 'HW_MAE': hw_mae, 'HW_R2': hw_r2
    }

    # -----------------------------------------
    # Grafik çizimi
    # -----------------------------------------
    plt.figure(figsize=(10, 6))
    plt.plot(y_test.index, y_test, label='Gerçek Değerler', color='blue', linestyle='--')
    plt.plot(y_test.index, sarima_preds, label='SARIMA Tahmin', color='red', linestyle=':')
    plt.plot(y_test.index, hw_preds, label='Holt-Winters Tahmin', color='orange')
    plt.title(f'{ulke} Kalkınma Endeksi Tahmin Sonuçları')
    plt.xlabel('Yıl')
    plt.ylabel('Kalkınma Endeksi')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# -----------------------------------------
# Sonuçları tablo olarak göster
# -----------------------------------------
sonuclar_df = pd.DataFrame(results).T
print(sonuclar_df)
```

## YAPAY SİNİR AĞLARI 750 EPOCH

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, Dropout, Input

# Veriyi hazırlama
data = en_iyi_df_sade.copy()
train = data[data['Yil'] <= 2010]  # 2000-2010 eğitim seti
test = data[data['Yil'] > 2010]   # 2011-2019 test seti

# Performans sonuçlarını saklamak için bir sözlük
results = {}

# Her ülke için modeli çalıştır ve sonuçları ayrı grafiklerde çiz
for ulke in data['Ulke'].unique():
    # Ülke bazında filtreleme
    train_ulke = train[train['Ulke'] == ulke].copy()
    test_ulke = test[test['Ulke'] == ulke].copy()

    # -----------------------------------------
    # ÖZELLİK MÜHENDİSLİĞİ
    # -----------------------------------------
    # Gecikmeli (Lag) ve Fark (Diff) özellikleri
    for lag in range(1, 4):  # 1-3 gecikme ekleyelim
        train_ulke[f'Lag{lag}'] = train_ulke['Kalkinma_Endeksi'].shift(lag)
        test_ulke[f'Lag{lag}'] = train_ulke['Kalkinma_Endeksi'].iloc[-lag]

    train_ulke['Diff'] = train_ulke['Kalkinma_Endeksi'].diff(1)
    test_ulke['Diff'] = test_ulke['Kalkinma_Endeksi'].diff(1)

    train_ulke['MA3'] = train_ulke['Kalkinma_Endeksi'].rolling(3).mean()
    test_ulke['MA3'] = test_ulke['Kalkinma_Endeksi'].rolling(3).mean()

    train_ulke.fillna(method='bfill', inplace=True)
    test_ulke.fillna(method='bfill', inplace=True)

    # -----------------------------------------
    # VERİ ÖLÇEKLENDİRME
    # -----------------------------------------
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()

    X_train = scaler_X.fit_transform(train_ulke[['Yil', 'Lag1', 'Lag2', 'Lag3', 'Diff', 'MA3']])
    X_test = scaler_X.transform(test_ulke[['Yil', 'Lag1', 'Lag2', 'Lag3', 'Diff', 'MA3']])

    y_train = scaler_y.fit_transform(train_ulke['Kalkinma_Endeksi'].values.reshape(-1, 1))
    y_test = scaler_y.transform(test_ulke['Kalkinma_Endeksi'].values.reshape(-1, 1))

    # 3D giriş formatına dönüştür
    X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    # -----------------------------------------
    # 1. LSTM Modeli
    # -----------------------------------------
    lstm_model = Sequential([
        Input(shape=(1, X_train.shape[2])),
        LSTM(50, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    lstm_model.compile(optimizer='adam', loss='mse')
    lstm_model.fit(X_train, y_train, epochs=750, batch_size=16, verbose=0)

    lstm_preds = scaler_y.inverse_transform(lstm_model.predict(X_test))

    # Performans değerlendirme
    lstm_rmse = np.sqrt(mean_squared_error(y_test, lstm_preds))
    lstm_mae = mean_absolute_error(y_test, lstm_preds)
    lstm_r2 = r2_score(y_test, lstm_preds)

    # -----------------------------------------
    # 2. RNN Modeli
    # -----------------------------------------
    rnn_model = Sequential([
        Input(shape=(1, X_train.shape[2])),
        SimpleRNN(50, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    rnn_model.compile(optimizer='adam', loss='mse')
    rnn_model.fit(X_train, y_train, epochs=750, batch_size=16, verbose=0)

    rnn_preds = scaler_y.inverse_transform(rnn_model.predict(X_test))

    # Performans değerlendirme
    rnn_rmse = np.sqrt(mean_squared_error(y_test, rnn_preds))
    rnn_mae = mean_absolute_error(y_test, rnn_preds)
    rnn_r2 = r2_score(y_test, rnn_preds)

    # -----------------------------------------
    # Sonuçları kaydet
    # -----------------------------------------
    results[ulke] = {
        'LSTM_RMSE': lstm_rmse, 'LSTM_MAE': lstm_mae, 'LSTM_R2': lstm_r2,
        'RNN_RMSE': rnn_rmse, 'RNN_MAE': rnn_mae, 'RNN_R2': rnn_r2
    }

    # Grafik çizimi
    plt.figure(figsize=(10, 6))
    plt.plot(test_ulke['Yil'], scaler_y.inverse_transform(y_test), label='Gerçek', color='blue')
    plt.plot(test_ulke['Yil'], lstm_preds, label='LSTM', color='green')
    plt.plot(test_ulke['Yil'], rnn_preds, label='RNN', color='orange')
    plt.title(f'{ulke} Tahmin Sonuçları')
    plt.legend()
    plt.show()

# Sonuçları tablo olarak göster
print(pd.DataFrame(results).T)
```
