```python
import pandas as pd

df = pd.read_csv('global_data.csv')
```

```python
df
```

```python
# Mevcut sütun isimlerini yeni isimlerle değiştirme
df.columns = [
    "Ulke",  # Entity
    "Yil",  # Year
    "Elektrik_Erisim_%",  # Access to electricity (% of population)
    "Temiz_Yakit_Erisim",  # Access to clean fuels for cooking
    "Yenilenebilir_Kapasite_KisiBasi",  # Renewable-electricity-generating-capacity-per-capita
    "Gelismis_Ulkeler_Finansal_Akisi",  # Financial flows to developing countries (US $)
    "Yenilenebilir_Enerji_Orani",  # Renewable energy share in the total final energy consumption (%)
    "Fosil_Yakit_Elektrik_TWh",  # Electricity from fossil fuels (TWh)
    "Nukleer_Elektrik_TWh",  # Electricity from nuclear (TWh)
    "Yenilenebilir_Elektrik_TWh",  # Electricity from renewables (TWh)
    "Dusuk_Karbon_Elektrik_%",  # Low-carbon electricity (% electricity)
    "KisiBasi_Enerji_Tuketimi",  # Primary energy consumption per capita (kWh/person)
    "Enerji_Yoğunluğu",  # Energy intensity level of primary energy (MJ/$2017 PPP GDP)
    "CO2_Emisyonu_kt",  # Value_co2_emissions_kt_by_country
    "Yenilenebilir_Enerji_%",  # Renewables (% equivalent primary energy)
    "Gdp_Buyume",  # gdp_growth
    "Gdp_KisiBasi",  # gdp_per_capita
    "Nufus_Yogunlugu",  # Density\n(P/Km2)
    "Yuzolcumu_Km2",  # Land Area(Km2)
    "Enlem",  # Latitude
    "Boylam"  # Longitude
]

# Yeni sütun isimlerini kontrol edelim
df
```

```python
missing_data = df.isnull().sum()
print("Eksik veri sayısı:", missing_data[missing_data > 0])
```

```python
# 'Nufus_Yogunlugu' sütununu float64 türüne çevirelim
df['Nufus_Yogunlugu'] = pd.to_numeric(df['Nufus_Yogunlugu'], errors='coerce')
```

```python
# Sadece sayısal sütunların seçilmesi
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Sadece sayısal sütunlardaki eksik değerleri sütunun ortalaması ile doldurulması
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

df.head(10)
```

```python
# 2020 yılına ait satırları çıkartma
df = df[df['Yil'] != 2020]
```

```python
df.info()
```

```python
df = df.copy()

df['Kalkinma_Endeksi'] = (-1 * (
    0.15 * df['Elektrik_Erisim_%'] +
    0.15 * df['Temiz_Yakit_Erisim'] +
    0.1 * df['Yenilenebilir_Kapasite_KisiBasi'] +
    0.1 * df['Yenilenebilir_Enerji_Orani'] +
    0.2 * df['Gdp_KisiBasi'] +
    0.15 * df['Gdp_Buyume'] -
    0.1 * df['Fosil_Yakit_Elektrik_TWh'] -
    0.05 * df['CO2_Emisyonu_kt']
    ))
```

```python
# Ülkelerin kalkınma endeksi ortalamalarını hesaplama
country_avg_index = df.groupby('Ulke')['Kalkinma_Endeksi'].mean().reset_index()

# Yeni veri seti: Sadece ülke adları ve ortalama kalkınma indeksleri
country_avg_index = country_avg_index.sort_values(by='Kalkinma_Endeksi', ascending=False)
```

```python
country_avg_index.tail(60)
```

```python
# En yüksek 4 ve en düşük 4 kalkınma endeksine sahip ülkeler
top_4_countries = country_avg_index.head(4)['Ulke']
bottom_4_countries = country_avg_index.tail(4)['Ulke']

# Bu ülkelerin veri setlerini almak
top_4_df = df[df['Ulke'].isin(top_4_countries)]
bottom_4_df = df[df['Ulke'].isin(bottom_4_countries)]

# Toplam veri seti
combined_df = pd.concat([top_4_df, bottom_4_df])
```
