# Korelasyon Analizi

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Veri seti (df_oecd) ve hedef değişkenin tanımlı olduğunu varsayıyoruz

# Özellikler ve hedef değişken
feature_columns = ['H1_Yoksulluk', 'H2_Aclik', 'H3_Saglik', 'H4_Egitim', 'H5_Esitsizlik', 
                   'H6_Su', 'H7_Enerji', 'H8_Is', 'H9_Altyapi', 'H10_Esitlik', 
                   'H11_Sehir', 'H12_Uretim', 'H13_Iklim', 'H14_SuHayati', 
                   'H15_KaraHayati', 'H16_Adalet', 'H17_Ortaklik']
target_column = 'SDG_Endeks'

# Hedef değişken ile özellikler arasındaki korelasyonları hesaplama
correlations = df_oecd[feature_columns + [target_column]].corr()[target_column].drop(target_column)

# Korelasyonları büyükten küçüğe sıralama
correlations_sorted = correlations.sort_values(ascending=False)

# En önemli 5 özelliği seçme
top_5_features = correlations_sorted.head(5)

print("Hedef değişkenle en yüksek korelasyona sahip 5 özellik:")
print(top_5_features)

# Korelasyon grafiği (heatmap) oluşturma
plt.figure(figsize=(10, 6))
sns.heatmap(df_oecd[top_5_features.index.tolist() + [target_column]].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("En Önemli Özelliklerin Korelasyonları")
plt.show()
```

```python
# Kullanılacak sütunlar (önemli hedefler)
selected_columns = ['H8_Is', 'H9_Altyapi', 'H16_Adalet', 'H10_Esitlik', 'H5_Esitsizlik']

# Her ülke için 2000-2022 ortalamalarını hesaplama
df_country_avg = df_oecd.groupby('Ulke')[selected_columns].mean().reset_index()

# Ortalamaları kontrol etme
print(df_country_avg.head())
```

## Kmeans Kümeleme

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# Elbow yöntemiyle optimal küme sayısını belirleme
inertia = []
silhouette_scores = []
calinski_scores = []
k_values = range(2, 11)  # 2'den 10'a kadar küme sayısını test etme

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, clusters))
    calinski_scores.append(calinski_harabasz_score(X_scaled, clusters))

# Elbow grafiği
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker="o", linestyle="--", color="b", label="Inertia")
plt.title("Elbow Yöntemi - Küme Sayısı")
plt.xlabel("Küme Sayısı")
plt.ylabel("Inertia (Hata Kareleri Toplamı)")
plt.grid(True)

# Silhouette Skoru grafiği
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker="o", linestyle="--", color="g", label="Silhouette Skoru")
plt.title("Silhouette Skoru - Küme Sayısı")
plt.xlabel("Küme Sayısı")
plt.ylabel("Silhouette Skoru")
plt.grid(True)

plt.tight_layout()
plt.show()

# Optimal küme sayısını belirleme (Elbow ve Silhouette Skoruna göre)
optimal_k = 3  # Grafikleri kontrol ederek manuel olarak belirlenir

# K-Means ile kümeleme
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme'] = clusters

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title(f"K-Means Kümeleme Sonuçları ({optimal_k} Küme)")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Kümeleme performans metrikleri
silhouette_avg = silhouette_score(X_scaled, clusters)
calinski_avg = calinski_harabasz_score(X_scaled, clusters)

print(f"Silhouette Skoru: {silhouette_avg:.2f}")
print(f"Calinski-Harabasz Skoru: {calinski_avg:.2f}")

# Küme dağılımlarını gösterme
print("Küme Dağılımları:")
print(df_country_avg['Küme'].value_counts())

# Her kümenin ortalama değerleri
print("\nHer Kümenin Özellik Ortalamaları:")
print(df_country_avg.groupby('Küme')[selected_columns].mean())

# Hangi kümede hangi ülkeler olduğunu gösterme
print("\nHangi Ülke Hangi Kümede:")
for cluster in sorted(df_country_avg['Küme'].unique()):
    print(f"\nKüme {cluster}:")
    countries_in_cluster = df_country_avg[df_country_avg['Küme'] == cluster]['Ulke'].tolist()
    print(", ".join(countries_in_cluster))
```

## DBSCAN kümeleme

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# DBSCAN modelini tanımlama
dbscan = DBSCAN(eps=1.4, min_samples=3)  # eps ve min_samples parametrelerini burada belirleyin
clusters = dbscan.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme'] = clusters

# Kümeleme performans metrikleri
valid_clusters = [cluster for cluster in set(clusters) if cluster != -1]  # -1'i çıkar
if len(valid_clusters) > 0:  # Performans metrikleri için geçerli kümeler mevcutsa
    silhouette_avg = silhouette_score(X_scaled[clusters != -1], clusters[clusters != -1])
    calinski_avg = calinski_harabasz_score(X_scaled[clusters != -1], clusters[clusters != -1])
    print(f"Silhouette Skoru: {silhouette_avg:.2f}")
    print(f"Calinski-Harabasz Skoru: {calinski_avg:.2f}")
else:
    print("DBSCAN modeli sadece bir küme buldu veya tüm noktalar gürültü olarak işaretlendi.")

# Küme dağılımlarını gösterme
print("\nKüme Dağılımları:")
print(df_country_avg['Küme'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("DBSCAN Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Her kümenin ortalama değerlerini hesaplama
if len(valid_clusters) > 0:  # Eğer geçerli kümeler varsa
    print("\nHer Kümenin Özellik Ortalamaları:")
    print(df_country_avg[df_country_avg['Küme'] != -1].groupby('Küme')[selected_columns].mean())
else:
    print("\nDBSCAN modeli geçerli küme bulamadı.")
```

## HDBSCAN  kümeleme

```python
import hdbscan
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# HDBSCAN modelini tanımlama ve kümeleme
hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=3, min_samples=3, gen_min_span_tree=True)
clusters = hdbscan_model.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme_HDBSCAN'] = clusters

# Gürültü (-1) kümeler hariç performans metriklerini hesaplama
if len(set(clusters)) > 1 and -1 in clusters:
    valid_clusters = clusters[clusters != -1]
    valid_X_scaled = X_scaled[clusters != -1]
    silhouette_avg = silhouette_score(valid_X_scaled, valid_clusters)
    calinski_avg = calinski_harabasz_score(valid_X_scaled, valid_clusters)
    print(f"HDBSCAN Silhouette Skoru: {silhouette_avg:.2f}")
    print(f"HDBSCAN Calinski-Harabasz Skoru: {calinski_avg:.2f}")
else:
    print("HDBSCAN modeli sadece bir küme veya gürültü buldu, performans metrikleri hesaplanamadı.")

# Küme dağılımlarını gösterme
print("\nHDBSCAN Küme Dağılımları:")
print(df_country_avg['Küme_HDBSCAN'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("HDBSCAN Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Her kümenin ortalama değerlerini hesaplama
if len(set(clusters)) > 1:  # Eğer birden fazla küme varsa
    print("\nHer Kümenin Özellik Ortalamaları:")
    print(df_country_avg[df_country_avg['Küme_HDBSCAN'] != -1].groupby('Küme_HDBSCAN')[selected_columns].mean())
else:
    print("\nHDBSCAN modeli sadece bir küme bulduğu için özellik ortalamaları hesaplanamadı.")
```
