## Sürdürülebilir Kalkınma Amaçları (SDGs)
Birleşmiş Milletler, 2015 yılında 2030 Sürdürülebilir Kalkınma Gündemi kapsamında 17 Küresel Amaç belirlemiştir. Bu amaçlar, dünya genelinde yoksulluk, eşitsizlik, iklim değişikliği ve çevresel bozulma gibi sorunları çözmeyi hedefler.

**17 Hedef ve Anlamları**

1.Yoksulluğa Son (No Poverty)

2.Açlığa Son (Zero Hunger)

3.Sağlık ve Kaliteli Yaşam (Good Health and Well-Being)

4.Nitelikli Eğitim (Quality Education)

5.Toplumsal Cinsiyet Eşitliği (Gender Equality)

6.Temiz Su ve Sanitasyon (Clean Water and Sanitation)

7.Erişilebilir ve Temiz Enerji (Affordable and Clean Energy)

8.İnsana Yakışır İş ve Ekonomik Büyüme (Decent Work and Economic Growth)

9.Sanayi, Yenilikçilik ve Altyapı (Industry, Innovation and Infrastructure)

10.Eşitsizliklerin Azaltılması (Reduced Inequality)

11.Sürdürülebilir Şehirler ve Topluluklar (Sustainable Cities and Communities)

12.Sorumlu Üretim ve Tüketim (Responsible Consumption and Production)

13.İklim Eylemi (Climate Action)

14.Sudaki Yaşam (Life Below Water)

15.Karasal Yaşam (Life on Land)

16.Barış, Adalet ve Güçlü Kurumlar (Peace, Justice and Strong Institutions)

17.Amaçlar için Ortaklıklar (Partnerships for the Goals)

```python
# OECD ülkeleri listesi
oecd_ulkeleri = [
    'Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Colombia',
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany',
    'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan',
    'Latvia', 'Lithuania', 'Mexico', 'Netherlands',
    'New Zealand', 'Norway', 'Luxembourg' ,'Poland', 'Portugal', 'Slovak Republic', 'Slovenia',
    'Spain', 'Sweden', 'Switzerland', 'United Kingdom', 'United States'
]

# OECD ülkelerine göre veri setini filtreleme
df_oecd = df[df['Ulke'].isin(oecd_ulkeleri)]

# Filtrelenmiş veri setini kontrol etme
df_oecd.reset_index(drop=True, inplace=True)
df_oecd
```

```python
# Belirli bir ülkenin verilerini görmek için bir fonksiyon yazalım

def ulke_verileri_goster(ulke_adi):
    """
    Belirtilen ülkenin veri setindeki tüm yıllarını ve verilerini gösterir.
    """
    ulke_verileri = df_oecd[df_oecd['Ulke'] == ulke_adi]
    if ulke_verileri.empty:
        print(f"{ulke_adi} için veri bulunamadı.")
    else:
        return ulke_verileri

# Test etmek için bir örnek ülke
ulke_verileri_goster("Mexico")  # Buraya görmek istediğiniz ülke adını yazabilirsiniz.
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Sonuçları saklamak için bir sözlük
country_results = {}

# Benzersiz ülkelerin listesi
countries = df_oecd['Ulke'].unique()

# Her bir ülke için işlemleri tekrarlama
for country in countries:
    print(f"Ülke: {country}")
    
    # Ülkenin verilerini filtreleme
    country_data = df_oecd[df_oecd['Ulke'] == country]
    
    # Özellikler ve hedef değişken
    feature_columns = ['H1_Yoksulluk', 'H2_Aclik', 'H3_Saglik', 'H4_Egitim', 'H5_Esitsizlik', 
                       'H6_Su', 'H7_Enerji', 'H8_Is', 'H9_Altyapi', 'H10_Esitlik', 
                       'H11_Sehir', 'H12_Uretim', 'H13_Iklim', 'H14_SuHayati', 
                       'H15_KaraHayati', 'H16_Adalet', 'H17_Ortaklik']
    X = country_data[feature_columns]
    y = country_data['SDG_Endeks']
    
    # Eğitim ve test setlerini yıllara göre ayırma
    train_years = range(2000, 2018)
    test_years = range(2018, 2023)
    X_train = X[country_data['Yil'].isin(train_years)]
    y_train = y[country_data['Yil'].isin(train_years)]
    X_test = X[country_data['Yil'].isin(test_years)]
    y_test = y[country_data['Yil'].isin(test_years)]
    
    # Özellik önem analizi için Random Forest Regressor
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)
    
    # Özellik önemlerini çıkarma
    feature_importances = rf_model.feature_importances_
    importance_df = pd.DataFrame({'Özellik': feature_columns, 'Önem': feature_importances})
    importance_df = importance_df.sort_values(by='Önem', ascending=False)
    
    # En önemli 5 özelliği seçme
    important_features = importance_df['Özellik'].head(5).tolist()
    
    # Yalnızca önemli özellikleri seçme
    X_cluster = country_data[important_features]
    
    # Veriyi ölçeklendirme
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_cluster)
    
    # K-Means kümeleme (3 küme varsayılmıştır, değiştirilebilir)
    kmeans = KMeans(n_clusters=3, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)
    
    # Küme etiketlerini ülke verisine ekleme
    country_data['Küme'] = clusters
    
    # Sonuçları kaydetme
    country_results[country] = {
        'Önemli Özellikler': important_features,
        'Küme Verisi': country_data[['Yil', 'Küme']],
        'Kümeleme': country_data['Küme'].value_counts()
    }
    
    # Küme sonuçlarını görselleştirme
    plt.figure(figsize=(10, 6))
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=100)
    plt.title(f"{country} - K-Means Kümeleme Sonuçları")
    plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
    plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
    plt.grid(True)
    plt.show()

# Tüm ülkeler için önemli özellikleri yazdırma
for country, result in country_results.items():
    print(f"Ülke: {country}")
    print(f"Önemli Özellikler: {result['Önemli Özellikler']}")
    print(result['Kümeleme'])
    print()
```

```python
# Kullanılacak sütunlar (önemli hedefler)
selected_columns = ['H3_Saglik', 'H9_Altyapi', 'H13_Iklim', 'H7_Enerji']

# Her ülke için 2000-2022 ortalamalarını hesaplama
df_country_avg = df_oecd.groupby('Ulke')[selected_columns].mean().reset_index()

# Ortalamaları kontrol etme
print(df_country_avg.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# Elbow yöntemiyle optimal küme sayısını belirleme
inertia = []
silhouette_scores = []
calinski_scores = []
k_values = range(2, 11)  # 2'den 10'a kadar küme sayısını test etme

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, clusters))
    calinski_scores.append(calinski_harabasz_score(X_scaled, clusters))

# Elbow grafiği
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker="o", linestyle="--", color="b", label="Inertia")
plt.title("Elbow Yöntemi - Küme Sayısı")
plt.xlabel("Küme Sayısı")
plt.ylabel("Inertia (Hata Kareleri Toplamı)")
plt.grid(True)

# Silhouette Skoru grafiği
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker="o", linestyle="--", color="g", label="Silhouette Skoru")
plt.title("Silhouette Skoru - Küme Sayısı")
plt.xlabel("Küme Sayısı")
plt.ylabel("Silhouette Skoru")
plt.grid(True)

plt.tight_layout()
plt.show()

# Optimal küme sayısını belirleme (Elbow ve Silhouette Skoruna göre)
optimal_k = 3  # Grafikleri kontrol ederek manuel olarak belirlenir

# K-Means ile kümeleme
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme'] = clusters

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title(f"K-Means Kümeleme Sonuçları ({optimal_k} Küme)")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Kümeleme performans metrikleri
silhouette_avg = silhouette_score(X_scaled, clusters)
calinski_avg = calinski_harabasz_score(X_scaled, clusters)

print(f"Silhouette Skoru: {silhouette_avg:.2f}")
print(f"Calinski-Harabasz Skoru: {calinski_avg:.2f}")

# Küme dağılımlarını gösterme
print("Küme Dağılımları:")
print(df_country_avg['Küme'].value_counts())

# Her kümenin ortalama değerleri
print("\nHer Kümenin Özellik Ortalamaları:")
print(df_country_avg.groupby('Küme')[selected_columns].mean())
```

## DBSCAN kümeleme

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# DBSCAN modelini tanımlama
dbscan = DBSCAN(eps=1, min_samples=3)  # eps ve min_samples parametrelerini burada belirleyin
clusters = dbscan.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme'] = clusters

# Kümeleme performans metrikleri
valid_clusters = [cluster for cluster in set(clusters) if cluster != -1]  # -1'i çıkar
if len(valid_clusters) > 0:  # Performans metrikleri için geçerli kümeler mevcutsa
    silhouette_avg = silhouette_score(X_scaled[clusters != -1], clusters[clusters != -1])
    calinski_avg = calinski_harabasz_score(X_scaled[clusters != -1], clusters[clusters != -1])
    print(f"Silhouette Skoru: {silhouette_avg:.2f}")
    print(f"Calinski-Harabasz Skoru: {calinski_avg:.2f}")
else:
    print("DBSCAN modeli sadece bir küme buldu veya tüm noktalar gürültü olarak işaretlendi.")

# Küme dağılımlarını gösterme
print("\nKüme Dağılımları:")
print(df_country_avg['Küme'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("DBSCAN Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Her kümenin ortalama değerlerini hesaplama
if len(valid_clusters) > 0:  # Eğer geçerli kümeler varsa
    print("\nHer Kümenin Özellik Ortalamaları:")
    print(df_country_avg[df_country_avg['Küme'] != -1].groupby('Küme')[selected_columns].mean())
else:
    print("\nDBSCAN modeli geçerli küme bulamadı.")
```

## Hiyerarşik Kümeleme

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# Agglomerative Clustering modelini tanımlama
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')  # Küme sayısını belirleyin
clusters = agg_clustering.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme'] = clusters

# Kümeleme performans metrikleri
silhouette_avg = silhouette_score(X_scaled, clusters)
calinski_avg = calinski_harabasz_score(X_scaled, clusters)

print(f"Silhouette Skoru: {silhouette_avg:.2f}")
print(f"Calinski-Harabasz Skoru: {calinski_avg:.2f}")

# Küme dağılımlarını gösterme
print("\nKüme Dağılımları:")
print(df_country_avg['Küme'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("Agglomerative Clustering Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Her kümenin ortalama değerlerini hesaplama
print("\nHer Kümenin Özellik Ortalamaları:")
print(df_country_avg.groupby('Küme')[selected_columns].mean())
```

## Birleşik Modelleme: KMeans - GMM

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# K-Means ile başlangıç kümeleme
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(X_scaled)

# Gaussian Mixture Model ile iyileştirme
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_clusters = gmm.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme_KMeans'] = kmeans_clusters
df_country_avg['Küme_GMM'] = gmm_clusters

# Kümeleme performans metrikleri
silhouette_kmeans = silhouette_score(X_scaled, kmeans_clusters)
silhouette_gmm = silhouette_score(X_scaled, gmm_clusters)
calinski_kmeans = calinski_harabasz_score(X_scaled, kmeans_clusters)
calinski_gmm = calinski_harabasz_score(X_scaled, gmm_clusters)

print(f"K-Means Silhouette Skoru: {silhouette_kmeans:.2f}")
print(f"K-Means Calinski-Harabasz Skoru: {calinski_kmeans:.2f}")
print(f"GMM Silhouette Skoru: {silhouette_gmm:.2f}")
print(f"GMM Calinski-Harabasz Skoru: {calinski_gmm:.2f}")

# Küme dağılımları
print("\nK-Means Küme Dağılımları:")
print(df_country_avg['Küme_KMeans'].value_counts())
print("\nGMM Küme Dağılımları:")
print(df_country_avg['Küme_GMM'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(12, 6))

# K-Means görselleştirme
plt.subplot(1, 2, 1)
scatter_kmeans = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=kmeans_clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("K-Means Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter_kmeans, label="Küme Etiketleri")
plt.grid(True)

# GMM görselleştirme
plt.subplot(1, 2, 2)
scatter_gmm = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=gmm_clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("GMM Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter_gmm, label="Küme Etiketleri")
plt.grid(True)

plt.tight_layout()
plt.show()

# Küme özellik ortalamaları
print("\nK-Means Küme Özellik Ortalamaları:")
print(df_country_avg.groupby('Küme_KMeans')[selected_columns].mean())
print("\nGMM Küme Özellik Ortalamaları:")
print(df_country_avg.groupby('Küme_GMM')[selected_columns].mean())
```

## HDBSCAN  kümeleme

```python
import hdbscan
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import pandas as pd

# Özellikleri ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_country_avg[selected_columns])

# HDBSCAN modelini tanımlama ve kümeleme
hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=3, min_samples=3, gen_min_span_tree=True)
clusters = hdbscan_model.fit_predict(X_scaled)

# Küme etiketlerini veri setine ekleme
df_country_avg['Küme_HDBSCAN'] = clusters

# Gürültü (-1) kümeler hariç performans metriklerini hesaplama
if len(set(clusters)) > 1 and -1 in clusters:
    valid_clusters = clusters[clusters != -1]
    valid_X_scaled = X_scaled[clusters != -1]
    silhouette_avg = silhouette_score(valid_X_scaled, valid_clusters)
    calinski_avg = calinski_harabasz_score(valid_X_scaled, valid_clusters)
    print(f"HDBSCAN Silhouette Skoru: {silhouette_avg:.2f}")
    print(f"HDBSCAN Calinski-Harabasz Skoru: {calinski_avg:.2f}")
else:
    print("HDBSCAN modeli sadece bir küme veya gürültü buldu, performans metrikleri hesaplanamadı.")

# Küme dağılımlarını gösterme
print("\nHDBSCAN Küme Dağılımları:")
print(df_country_avg['Küme_HDBSCAN'].value_counts())

# Kümeleme sonuçları için görselleştirme
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_scaled[:, 0], X_scaled[:, 1], 
    c=clusters, cmap='viridis', s=100, alpha=0.7, edgecolor='k'
)
plt.title("HDBSCAN Kümeleme Sonuçları")
plt.xlabel("Özellik 1 (Ölçeklendirilmiş)")
plt.ylabel("Özellik 2 (Ölçeklendirilmiş)")
plt.colorbar(scatter, label="Küme Etiketleri")
plt.grid(True)
plt.show()

# Her kümenin ortalama değerlerini hesaplama
if len(set(clusters)) > 1:  # Eğer birden fazla küme varsa
    print("\nHer Kümenin Özellik Ortalamaları:")
    print(df_country_avg[df_country_avg['Küme_HDBSCAN'] != -1].groupby('Küme_HDBSCAN')[selected_columns].mean())
else:
    print("\nHDBSCAN modeli sadece bir küme bulduğu için özellik ortalamaları hesaplanamadı.")
```
