## Bu aşamadan sonrası test için yazılmıştır. (düzenlenecektir)

## Afganistan için yıllık bazda kişi başı enerji tüketim verisi üzerinde zaman seri analizi kullanılarak ülkelerin sürdürülebilir enerji kalkınma planları hakkında yorum yapılması hedeflenmiştir.(değiştirilecek)

```python
print(df.columns)
```

```python
# Yıllık verilere göre kişi başı enerji tüketimi sütununu seçelim
df['Yil'] = pd.to_datetime(df['Yil'], format='%Y')
df.set_index('Yil', inplace=True)

# İlgili sütunları seçmek için örnek (ülke seçimi için bir filtre ekleyelim)
ulke = 'Angola'
ulke_df = df[df['Ulke'] == ulke][['KisiBasi_Enerji_Tuketimi']]
```

```python
# Zaman serisi çizimi
plt.figure(figsize=(10, 6))
plt.plot(ulke_df.index, ulke_df['KisiBasi_Enerji_Tuketimi'], marker='o')
plt.title(f"{ulke} - Kişi Başı Enerji Tüketimi")
plt.xlabel('Yıl')
plt.ylabel('Kişi Başı Enerji Tüketimi')
plt.grid(True)
plt.show()
```

 ## Durağanlık testi

```python
# Augmented Dickey-Fuller (ADF) testi
adf_test = adfuller(ulke_df['KisiBasi_Enerji_Tuketimi'])

print(f'ADF Test İstatistiği: {adf_test[0]}')
print(f'p-değeri: {adf_test[1]}')
if adf_test[1] < 0.05:
    print("Zaman serisi durağandır.")
else:
    print("Zaman serisi durağan değildir, fark alma işlemi yapılmalıdır.")
```

ADF test sonuçlarına göre, p-değeri 0.05’ten büyük olduğu için zaman serinizin durağan olmadığı sonucuna vardık. Zaman serilerini analiz ederken, durağan olmayan serilerde genellikle fark alma işlemi (differencing) uygulanarak serinin durağan hale getirilmesi gerekir.

Fark Alma İşlemi Nasıl Yapılır?
Zaman serilerindeki fark alma işlemi, önceki değeri çıkartarak serinin varyans ve ortalama değerlerini sabitlemeyi amaçlar. Bunu uygulayarak seriyi durağan hale getirebiliriz.

### 1. Birinci Dereceden Fark Alma (Differencing):

```python
# Birinci dereceden fark alma işlemi
ulke_df['KisiBasi_Enerji_Tuketimi_Diff'] = ulke_df['KisiBasi_Enerji_Tuketimi'].diff()

# Fark alma sonrası oluşan seriyi görselleştirme
plt.figure(figsize=(10, 6))
plt.plot(ulke_df.index, ulke_df['KisiBasi_Enerji_Tuketimi_Diff'], marker='o')
plt.title(f"{ulke} - Birinci Dereceden Fark Alınmış Kişi Başı Enerji Tüketimi")
plt.xlabel('Yıl')
plt.ylabel('Kişi Başı Enerji Tüketimi Farkı')
plt.grid(True)
plt.show()
```

### 2. ADF Testini Tekrar Uygulayın:

```python
# Fark alınmış veri üzerinde ADF testi uygulama
adf_test_diff = adfuller(ulke_df['KisiBasi_Enerji_Tuketimi_Diff'].dropna())

print(f'ADF Test İstatistiği (Fark Alınmış): {adf_test_diff[0]}')
print(f'p-değeri (Fark Alınmış): {adf_test_diff[1]}')
if adf_test_diff[1] < 0.05:
    print("Fark alınmış zaman serisi durağandır.")
else:
    print("Fark alınmış zaman serisi hala durağan değildir.")
```

 ### Açıklama:
Birinci Dereceden Fark Alma: Zaman serisindeki fark alma işlemi, önceki değeri çıkartarak yeni bir seriye dönüştürür. Bu işlem, durağan olmayan serilerin durağan hale gelmesine yardımcı olur.
ADF Testinin Tekrar Uygulanması: Fark alma işlemi sonrasında, serinin durağan olup olmadığını tekrar kontrol ederiz. p-değerinin 0.05’ten küçük olması, serinin artık durağan olduğunu gösterir.
Eğer fark alma işlemi sonrasında seriniz hala durağan değilse, ikinci veya daha yüksek dereceden fark alma işlemi de yapabilirsiniz.

# Model Seçimi (Auto-ARIMA ile Otomatik Parametre Seçimi)

```python
# Auto-ARIMA ile en iyi parametrelerin belirlenmesi
auto_arima_model = auto_arima(ulke_df['KisiBasi_Enerji_Tuketimi'], seasonal=False, trace=True)
print(auto_arima_model.summary())
```

Genel Değerlendirme
Seçilen model olan ARIMA(1,1,0), kişi başı enerji tüketimi verisi için uygun görünüyor.
Ljung-Box ve Jarque-Bera testlerinin sonuçları, modelin kalıntılarının otokorelasyon içermediğini ve normal dağılıma uyduğunu gösteriyor.
Ancak, heteroskedasticity testinde değişen varyans olduğu görülüyor, bu da modelleme sonuçlarını ve tahmin doğruluğunu etkileyebilir.

 # Model eğitimi ve Tahmin yapma

```python
# En iyi ARIMA modelini eğitme
train_size = int(len(ulke_df) * 0.8)
train, test = ulke_df.iloc[:train_size], ulke_df.iloc[train_size:]

# Yalnızca hedef sütunu seçiyoruz
train_target = train['KisiBasi_Enerji_Tuketimi']
test_target = test['KisiBasi_Enerji_Tuketimi']

# Modeli kurma
model = SARIMAX(train_target, order=auto_arima_model.order)
arima_model = model.fit(disp=False)

# Tahminler
predictions = arima_model.get_forecast(steps=len(test_target))
pred_mean = predictions.predicted_mean

# Tahmin sonuçlarını görselleştirin
plt.figure(figsize=(12, 6))
plt.plot(train.index, train_target, label='Eğitim Verisi')
plt.plot(test.index, test_target, label='Gerçek Değerler', color='orange')
plt.plot(test.index, pred_mean, label='Tahminler', color='green', linestyle='--')
plt.title(f"{ulke} - Kişi Başı Enerji Tüketimi Tahminleri")
plt.xlabel('Yıl')
plt.ylabel('Kişi Başı Enerji Tüketimi')
plt.legend()
plt.grid(True)
plt.show()
```

## Model performansını değerlendirme

```python
# Tahmin performansı değerlendirme metrikleri
mse = mean_squared_error(test['KisiBasi_Enerji_Tuketimi'], pred_mean)
mae = mean_absolute_error(test['KisiBasi_Enerji_Tuketimi'], pred_mean)
print(f"Ortalama Kare Hata (MSE): {mse}")
print(f"Ortalama Mutlak Hata (MAE): {mae}")
```
